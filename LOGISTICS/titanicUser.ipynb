```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
File = pd.read_csv('titanic.csv')
```

```python
# Storing Age column to see difference
Age = pd.Series(File['Age'])

# Print File
File
```

## Functions

    -> One-hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder


File['Sex'] = File['Sex'].astype('category')
a = pd.get_dummies(File['Sex']).astype(int)
File = File.join(a)

```

    -> Label Encoder

```python
encoder = LabelEncoder()
File['Embarked'] = encoder.fit_transform(File['Embarked'])
File
```

    -> Dropping Columns

```python
# File = File.drop(['PassengerId','Sex'],axis=1)
File = File.drop([0,1,6,7,8])
File
```

    -> Z-score Scaler

```python
Mean = File['Age'].mean()
StDev = File['Age'].std()

zScore = pd.Series((File['Age'] - Mean)/StDev , name='Z-Score ( Age )')
File = File.join(zScore)
plt.hist(zScore)
plt.title('Z-Score of Age')
File
```

    -> Min-Max Scaler

```python
Min = File['Age'].min()
Max = File['Age'].max()

MMScaler = pd.Series((File['Age'] - Min)/(Max-Min),name = 'Min-Max Scaler ( age )')
File = File.join(MMScaler)
plt.hist(MMScaler)
plt.title('Min - Max Scaler of Age')
File
```

    -> IQR

```python
Q1 = File['Age'].quantile(0.25)
Q3 = File['Age'].quantile(0.75)
IQR = Q3-Q1
```

```python
# Lower Fence
LowerFence = Q1 - 1.5 * IQR
# Upper Fence
UpperFence = Q3 + 1.5 * IQR
```

    -> Removing Outliers

```python
File = File.query("Age > @LowerFence and Age < @UpperFence")
plt.subplot(1,2,1)
plt.hist(File['Age'])
plt.title('Age (Before)')
plt.subplot(1,2,2)
plt.hist(Age)
plt.title('Age (after)')
```

#### Model Training


```python
#Import Modules
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

    -> Getting Dataframe Ready

```python
NewDataFrame = pd.DataFrame(File)
NewDataFrame
```

```python
NewDataFrame = pd.DataFrame(File)
NewDataFrame = NewDataFrame.drop(['PassengerId','Z-Score ( Age )','Name','Min-Max Scaler ( age )','Sex','Ticket','Cabin'],axis = 1)
NewDataFrame
```

    -> Dataset Split

```python
X = NewDataFrame.loc[:,NewDataFrame.columns != 'Survived']
y = NewDataFrame['Survived']

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
# Print Splitted Data
print(x_train,x_test,y_train,y_test)
```

    -> Logistic Regression

```python
regressionFunc = LogisticRegression()
regressionFunc.fit(x_train,y_train)
```

    -> Accuracy Of the Model

```python
# Accuracy Prediction
y_pred = regressionFunc.predict(x_test)
print(accuracy_score(y_test,y_pred)*100)

```

    -> Regression model Line

```python
print("Co efficient : ",regressionFunc.coef_)
print("Intercept : ",regressionFunc.intercept_)
```
